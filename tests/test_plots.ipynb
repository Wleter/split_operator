```python
from matplotlib import animation
import numpy as np
import matplotlib.pyplot as plt

def wave_animation(prefix, grid_name, xlabel, with_potential=False): 
    wave = np.load(f'test_data/{prefix}.npy')
    grid = np.load(f'test_data/{prefix}_{grid_name}_grid.npy')
    if with_potential:
        potential = np.load(f'test_data/{prefix}_potential.npy')

    fig, ax = plt.subplots()

    def animate(i):
        ax.clear()
        ax.plot(grid, wave[:, i])
        if with_potential:
            ax.plot(grid, potential * np.max(wave[:, i]) / np.max(potential), color='black')
            
        ax.set_xlabel(xlabel)
        ax.set_ylabel("wave function density")

    anim = animation.FuncAnimation(fig, animate, interval=60, frames=wave.shape[1], blit=False)
    return anim
```

```python
anim = wave_animation("harmonic", "space", "distance [bohr]", True)
anim.save('test_plots/harmonic_animation.gif', writer="pillow", fps=30)
```
